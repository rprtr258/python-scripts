```
from math import atan, pi, sin

import matplotlib.pyplot as plt
```

```
def to_atan(x):
    return atan(x) / (pi / 2)
```

```
def plot_atan(f, xdeg=1, ydeg=1, drawone=False):
    xs = sorted([x / 100 for x in range(-100000, 100000)] + [(x / 10000) ** 1 for x in range(-100000, 100000)])
    fs = [f(x) for x in xs]

    axs = [to_atan(x) ** xdeg for x in xs]
    afs = [to_atan(y) ** ydeg for y in fs]

    if drawone:
        f2s = [1 for x in xs]
        af2s = [to_atan(y) ** ydeg for y in f2s]

    plt.figure(figsize=(12, 12))

    SHIT = 1.005
    plt.plot([0, 0], [-SHIT, SHIT], c='k', linewidth=1)
    plt.plot([-SHIT, SHIT], [0, 0], c='k', linewidth=1)
    plt.plot([-SHIT, SHIT], [-SHIT, SHIT], c='k', linewidth=1)
    plt.plot([-SHIT, SHIT], [SHIT, -SHIT], c='k', linewidth=1)
    plt.plot(axs, afs)
    if drawone:
        plt.plot(axs, af2s)

    plt.xlim(-SHIT, SHIT)
    plt.ylim(-SHIT, SHIT)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```
from math import cos

plot_atan(lambda x: sin(x) * x * cos(x))
```

```
plot_atan(lambda x: sin(x) * x, xdeg=3, ydeg=3)
```

```
plot_atan(lambda x: sin(1 / x) if x != 0.0 else 0)
```

```
plot_atan(lambda x: x * sin(1 / x) if x != 0.0 else 0, drawone=True)
```

```
cache = {}
def weierstrass(x, N=100):
    if x in cache:
        return cache[x]
    cache[x] = 0
    for n in range(1,N):
        cache[x] += cos(3**n*pi*x)/2**n
    return cache[x]

plot_atan(lambda x: weierstrass(x))
```

```
from math import floor

plot_atan(lambda x: x - floor(x))
```

```
plot_atan(lambda x: floor(x))
```

```
def plot_polar_curve_atan(f, xdeg=1, ydeg=1):
    ts = sorted([x / 100 for x in range(1, 100000)] + [(x / 10000) ** 1 for x in range(1, 100000)])
    fs = [(r * cos(phi), r * sin(phi)) for r, phi in map(f, ts)]

    axs = [to_atan(x) ** xdeg for x, _ in fs]
    ays = [to_atan(y) ** ydeg for _, y in fs]

    plt.figure(figsize=(12, 12))

    SHIT = 1.005
    plt.plot([0, 0], [-SHIT, SHIT], c='k', linewidth=1)
    plt.plot([-SHIT, SHIT], [0, 0], c='k', linewidth=1)
    plt.plot([-SHIT, SHIT], [-SHIT, SHIT], c='k', linewidth=1)
    plt.plot([-SHIT, SHIT], [SHIT, -SHIT], c='k', linewidth=1)
    plt.plot(axs, ays)

    plt.xlim(-SHIT, SHIT)
    plt.ylim(-SHIT, SHIT)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```
plot_polar_curve_atan(lambda t: (t, t), xdeg=13, ydeg=13)
```

```
plot_polar_curve_atan(lambda t: (t * (sin(t) + 1), t))
```

```
from math import tanh

plot_polar_curve_atan(lambda t: (weierstrass(t) + 1, tanh(t) * pi * t))
```

```
plot_polar_curve_atan(lambda t: (weierstrass(t) + 1, t * pi))
```

```
ts = sorted([x / 100 for x in range(1, 100000)] + [(x / 10000) ** 1 for x in range(1, 100000)])
fs = [(r * cos(phi), r * sin(phi)) for r, phi in [(weierstrass(t) + 1, t * pi) for t in ts]]

axs = [x for x, _ in fs]
ays = [y for _, y in fs]

plt.figure(figsize=(12, 12))

plt.plot(axs, ays)

plt.xticks([])
plt.yticks([])
plt.show()
```
